```python
# 전국단위 전기차 대수 집계
import pandas as pd
```

```python
df = pd.read_excel("2022년_09월_자동차_등록자료_통계.xlsx",engine="openpyxl",sheet_name="12.배기별(승용차)_등록현황",header=2)
df.head()
```

```python
df.rename(columns={'Unnamed: 0':'시도'},inplace=True)
```

```python
df.head()
```

```python
df1 = df[['시도']]
df1.dropna()
```

```python
df2 = df[['저속전기차','전기차']].sum(axis=1)
df2
```

```python
res = pd.concat([df1,df2],axis=1)
res
```

```python
arr = []
name = []
cp = res.loc[1] + res.loc[4]+ res.loc[9]
arr.append(cp)
name.append("수도권")
gs = res.loc[2] + res.loc[3] + res.loc[7] + res.loc[15] + res.loc[16]
arr.append(gs)
name.append("경상도")
jl = res.loc[5] + res.loc[13] + res.loc[14]
arr.append(jl)
name.append("전라도")
cc = res.loc[6] + res.loc[8] + res.loc[11] + res.loc[12]
arr.append(cc)
name.append("충청도")
gw = res.loc[10]
arr.append(gw)
name.append("강원도")
jj = res.loc[17]
arr.append(jj)
name.append("제주도")
sum = res.loc[18]
arr.append(sum)
name.append("총합")
```

```python
for i in range(len(arr)):
    arr[i]=arr[i].set_axis(labels=[0,name[i]])
```

```python
df_res = pd.concat([arr[0],arr[1],arr[2],arr[3],arr[4],arr[5],arr[6]])
df_res = pd.DataFrame(df_res)
```

```python
df_res
```

```python
df_res = df_res.drop([0])
```

```python
df_res
```

```python
df_res2 = df_res.div(df_res.loc["총합",0])
```

```python
df_res2 = df_res2*100
```

```python
df_res2 = df_res2.rename(columns={0:"a"})
```

```python
df_res2
```

```python
result = pd.concat([df_res,df_res2], axis=1)
```

```python
result
```

```python
result.to_csv("numevcar.csv",header=None)
```
