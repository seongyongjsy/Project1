```python
import pandas as pd
```

```python
df = pd.read_excel("2022년_09월_자동차_등록자료_통계.xlsx",engine="openpyxl",sheet_name="04.성별_연령별",header=2)
df = df[['성별','연령/시도','총계']]
df
```

```python
num_1 = df.query('성별=="여성"').index.values[0]
df_man = df[:num_1-1]
df_man = df_man.rename(columns={'연령/시도':'age','총계':'num_man'})
```

```python
num_2 = df.query('성별=="기타"').index.values[0]
df_women = df[num_1:num_2-1]
df_women = df_women.reset_index(drop=True)
df_women = df_women.rename(columns={'연령/시도':'age','총계':'num_women'})
```

```python
df_man = df_man.drop(columns='성별')
df_women = df_women.drop(columns='성별')
```

```python
df_man, df_women
```

```python
df_man = df_man.set_index(keys=['age'],drop=True)
df_women = df_women.set_index(keys=['age'],drop=True)
```

```python
res = pd.concat([df_man,df_women],axis=1)
res.columns = ['num_man','num_women']
res
```

```python
res.to_csv("numcar.csv",header=None)
```
