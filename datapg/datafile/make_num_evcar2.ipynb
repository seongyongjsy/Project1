```python
df = df.drop([0])
df = df[:8]
df.rename(columns={'Unnamed: 0':'시도'},inplace=True)
df = df[["시도","저속전기차","전기차"]]
df
```

```python
df1 = df[["시도"]]
df2 = df.sum(axis=1)
# 각 광역시 별 데이터프레임에 추가 + 전체값 반영해서 퍼센트 열 추가

df_res = pd.concat([df1,df2],axis=1)
df_res
```

```python
df_pl1 = df_res.loc[1] # 서울
df_pl1 = pd.DataFrame(df_pl1)

df_pl2 = df_res.loc[4] # 인천
df_pl2 = pd.DataFrame(df_pl2)

df_plgg = pd.concat([df_pl1.T,df_pl2.T]) # 수도권에 서울값 인천값 합치기위해 데이터프레임 생성
df_plgg.rename(columns={'시도':'시군별(2)',0:"전기"},inplace=True)
df_plgg
```

```python
newdfgg = pd.concat([df_plgg,dfgg])
newdfgg = newdfgg.reset_index(drop=True)
newdfgg
```

```python
newdfgg.loc[2,"전기"] = newdfgg.loc[0:2].sum()[1]
newdfgg = newdfgg.sort_values("전기",ascending=False)
newdfgg = newdfgg[:7]

newdfgg2 = newdfgg[["전기"]] # 비율을 구하기 위해 전기차 대수 값만 추출
newdfgg2 =  newdfgg2.div(newdfgg.loc[2,"전기"])
newdfgg2 = newdfgg2*100 # 전체로 나누고 100을 곱해서 완료
newdfgg2.rename(columns={'전기':'비율'},inplace=True)
newdfgg2
```

```python
realdfgg = pd.concat([newdfgg,newdfgg2],axis=1) # 경기도 값이 쨘!
realdfgg
```
